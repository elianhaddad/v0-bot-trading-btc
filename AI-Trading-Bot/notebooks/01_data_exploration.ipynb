
# üìä Data Exploration of BTC/USDT Prices

In this notebook, we will:
- Load BTC/USDT data collected from Binance (for 2025 or any date range you choose).
- Check for missing data.
- Explore trading volume.
- Analyze volatility.
- Visualize price distribution.
- Summarize observations.

This step will help us ensure our data is clean and understand if 1-minute data is sufficient for intraday AI trading.


```python

import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV (assuming data_collector.py fetched 2025 data)
df = pd.read_csv('../data/BTCUSDT_1m.csv')

df.tail(50)
```


## üõ†Ô∏è Preprocessing Data

- Convert `open_time` to datetime.
- Set it as the index.
- Convert numeric columns for analysis.


```python

df['open_time'] = pd.to_datetime(df['open_time'])
df.set_index('open_time', inplace=True)

# Convert 'close' and 'volume' to numeric
df['close'] = pd.to_numeric(df['close'], errors='coerce')
df['volume'] = pd.to_numeric(df['volume'], errors='coerce')

df.info()
```


## üîé Checking Missing Data


```python

missing_values = df.isnull().sum()
print('Missing values per column:')
print(missing_values)
```


## üìà Basic Statistics


```python

df.describe()
```


## üíπ Visualizing BTC Price Over Time


```python

plt.figure(figsize=(14, 7))
df['close'].plot(title='BTC/USDT Price (1-minute intervals)', grid=True)
plt.xlabel('Date')
plt.ylabel('BTC Price (USD)')
plt.show()
```


## üìä Analyzing Trading Volume


```python

plt.figure(figsize=(14, 7))
df['volume'].plot(title='BTC/USDT Trading Volume (1-minute intervals)', grid=True, color='orange')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```


## üìâ Volatility Analysis

We'll calculate the minute-to-minute returns to see how volatile the price is.


```python

df['returns'] = df['close'].pct_change()

# Plot returns
plt.figure(figsize=(14, 7))
df['returns'].plot(title='BTC/USDT Returns (minute-to-minute)', grid=True, color='green')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.show()

# Describe returns
df['returns'].describe()
```


## üßÆ Price Distribution


```python

plt.figure(figsize=(10, 6))
df['close'].hist(bins=100, color='skyblue')
plt.title('Distribution of BTC/USDT Prices')
plt.xlabel('BTC Price')
plt.ylabel('Frequency')
plt.show()
```
